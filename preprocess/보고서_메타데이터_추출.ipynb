### 옵션1: marker-pdf

```python
# pip install marker-pdf
# 새로운 가상환경에서 실행 권장 - 과도하게 많은 라이브러리 강제 업데이트 (ex. 파이토치 최신버전 강제 업데이트)
```

```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

converter = PdfConverter(
    artifact_dict=create_model_dict(),
)
```

```python
rendered = converter("지속가능경영_보고서_2024.pdf")
```

```python
text, _, images = text_from_rendered(rendered)
```

```python
with open("./marker-pdf/_content.txt", "w") as file:
    file.writelines(text)
```

```python
for filename, img in images.items():
    img.save("./marker-pdf/" + filename)
```

```python
# 클리어링하고 새로저장
with open("./marker-pdf/_content.txt", "r", encoding="utf-8") as file:
    content = file.read()

import re
modified_content = re.sub("<br>", " ", content)
modified_content = re.sub("</sup>", "", modified_content)
modified_content = re.sub("<sup>", "", modified_content)
modified_content = re.sub("</sub>", "", modified_content)
modified_content = re.sub("<sub>", "", modified_content)

with open("./marker-pdf/_content_v1.txt", "w", encoding="utf-8") as file:
    file.write(modified_content)
```

### 옵션2: docling

```python
# pip install docling
```

```python
from docling.document_converter import DocumentConverter
converter = DocumentConverter()
```

```python
source = "지속가능경영_보고서_2024.pdf"  # document per local path or URL
result = converter.convert(source)
```

```python
print(result.document.export_to_markdown())  # output: "## Docling Technical Report[...]"
```
